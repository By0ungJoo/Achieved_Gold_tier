```python
# KDA(Kill Assist Death) 값에 따른 승리 그래프 
# KDA 컬럼 추가  KDA = (kill + assist) / death _ 보통 3이상부터 게임안에서 1인분 이상한 것으로 봄
df['kda'] = round((df['kills'] + df['assists']) / df['deaths'], 1)

# kda 컬럼에서도 켤측치 확인 
df['kda'].isnull()

# 왜 결측치가 나왔는지 확인
df[df['kda'].isnull() == True]

# 킬, 데스, 어시스트 모두 0이라 결측치가 되어서 0으로 대체 
df['kda'] = df['kda'].fillna(0)

```

```python
# 승리시에 높은 kda가 분포되어 있는걸 확인 
sns.scatterplot(x = 'win', y = 'kda', data = df)
plt.xlim(-1, 2)
plt.title('kda에 따른 승패')
```

```python
# 포지션에 따른 승률 비교
# 리그오브레전드에는 5개의 포지션이 존재하는데 크게 2가지로 나눌수 있다. 공격로 3군데를 가는 라이너, 정글로 들어가 중립몬스터를 사냥하는 정글러 
# 크게 나눈 두 부류는 스펠의 차이를 대표적으로 알수 있다. '강타'라는 스펠의 유무로 라이너 또는 정글러로 구분할 수 있다. 
#

print('강타 O ')
print(df[df['spell1Id'] == '강타']['win'].value_counts())
print(df[df['spell2Id'] == '강타']['win'].value_counts())

print('강타 X ')
print(df[df['spell2Id'] != '강타']['win'].value_counts())
print(df[df['spell1Id'] != '강타']['win'].value_counts())
```

```python
smite_o = pd.concat([df[df['spell1Id'] == '강타']['win'], df[df['spell2Id'] == '강타']['win']], axis = 0)
smite_x = pd.concat([df[df['spell1Id'] != '강타']['win'], df[df['spell2Id'] != '강타']['win']], axis = 0)
```

```python
smite_o = []
smite_x = []
for i in range(700):
    if df['spell1Id'][i] != '강타':
        if df['spell2Id'][i] != '강타':
            smite_x.append(df['win'][i])
        elif df['spell2Id'][i] == '강타':
            smite_o.append(df['win'][i])
            
    elif df['spell1Id'][i] == '강타':
        if df['spell2Id'][i] != '강타':
            smite_o.append(df['win'][i])
        else:
            pass
            


# 총 진행한 700판 중 정글러 포지션 225게임 그외 라이너 475게임 확인
print(len(smite_o))
print(len(smite_x))
```
