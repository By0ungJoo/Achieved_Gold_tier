```python
df['simple_length'] = None
for i in range(len(df)):
    if int(re.findall(r'\d+', df['game_length'][i])[0]) <= 20:
        df['simple_length'][i] = '초반'
#         print('초반')
    elif int(re.findall(r'\d+', df['game_length'][i])[0]) > 20 and int(re.findall(r'\d+', df['game_length'][i])[0]) <= 40:
        df['simple_length'][i] = '중반'
#         print('중반')
    else:
        df['simple_length'][i] = '후반'
#         print('후반')
```

```python
df.head()
```

```python
# 승리, 패배별 횟수
df1 = df.groupby(['simple_length', 'result']).size()
df1 = df1.reset_index()
df1 = df1.reindex(index=[2, 3, 0, 1, 4, 5])
df1.reset_index(drop = True)
df1 = df1.rename(columns={0: 'cnt'})
```

```python
# 승률
df1['win_rate'] = ''

for i in range(len(df1)):
    if i%2 == 0:
        df1['win_rate'][i] = round((df1['cnt'][i] / (df1['cnt'][i] + df1['cnt'][i+1])) * 100, 1)
    else:
        pass
    
df1
```

```python
plt.figure(figsize=(15,7))

plt.subplot(1, 2, 1)
bar_width = 0.35
alpha = 0.5
index = np.arange(3)
label = df1['simple_length'].unique()

p1 = plt.bar(index, df1[df1['result'] == '승리']['cnt'],
        bar_width,
        color = 'b',
        label = 'Win',
        alpha = alpha)

p2 = plt.bar(index + bar_width, df1[df1['result'] == '패배']['cnt'],
        bar_width,
        color = 'r',
        label = 'Lose',
        alpha = alpha)

plt.title('Result of Game_Length', fontsize=20)
plt.xlabel('Game_Length', fontsize=18)
plt.ylabel('Cnt', fontsize=18)
plt.xticks(index + (bar_width/2), label, fontsize=15)
plt.legend((p1[0], p2[0]), ('Win', 'Lose'), fontsize=15)
plt.grid()

plt.subplot(1, 2, 2)
plt.scatter(list(df1[df1['win_rate'] != '']['win_rate']), list(df1[df1['win_rate'] != '']['win_rate']),
            marker='.',
            s=200,
            color= 'b', 
            alpha= alpha)
plt.xticks(df1[df1['win_rate'] != '']['win_rate'], label, fontsize=15)
plt.xlim(40, 55)
plt.ylim(40, 55)
plt.grid()
plt.title('Win_rate Change', fontsize=20)
```

```python
# 초반 : 44.9% / 중반 : 47.8% / 후반 : 50% 
# 데이터가 많지 않지만 게임 시간에 따른 승률로 보았을 때 게임 시간이 길어질수록 승률이 좋아지는 걸 볼수 있다. 
```

```python
df3 = df.groupby(['result', 'op_score']).size()
df3 = df3.reset_index()
df3 = df3.rename(columns={0 : 'cnt'})
df3
```

```python
plt.figure(figsize=(12, 8))
alpha = 0.5

p1 = plt.scatter(df3[df3['result'] == '승리']['op_score'], df3[df3['result'] == '승리']['cnt'],
                color = 'b', 
                alpha= alpha)

p2 = plt.scatter(df3[df3['result'] == '패배']['op_score'], df3[df3['result'] == '패배']['cnt'],
                color = 'r',
                alpha= alpha)

plt.grid()
plt.title('Win_rate Change', fontsize=20)
plt.xlabel('OP_score', fontsize = 15)
plt.xlim(2, 11)
plt.ylim(0, 15)
```

```python
# 평균 op_score 점수 확인 : 6.188520971302426

d = 0 

for i in range(len(df3)):
    d = d + (df3['op_score'][i] * df3['cnt'][i])
    
d / df3['cnt'].sum()
```

```python
# OP score에 따른 게임 결과는 승리와 패배가 분포되어 있는 상태가 비슷
# 6.1 점대가 이 계정의 평균 확인 
# op_score가 높아질수록 승리의 횟수가 많아지고 반대로 낮아질수록 패배의 횟수도 많아지는 것 확인 
```

```python
df3[df3['op_score'] >= 6.1]['cnt'].sum()
```

```python
df_win = df3[df3['result'] == '승리']
df_lose = df3[df3['result'] == '패배']
```

```python
print(df_lose['cnt'].sum())
print(df_lose[df_lose['op_score'] < 6.1]['cnt'].sum())
```

```python
# 승리한 게임 중 약 75%정도는 op_score 평균치 이상 기록 
print(str(round(df_win[df_win['op_score'] > 6.1]['cnt'].sum() / df_win['cnt'].sum() * 100)) + '%')

# 패배한 게임 중 약 68%는 op_score 평균치 이하 기록 
print(str(round(df_lose[df_lose['op_score'] < 6.1]['cnt'].sum() / df_lose['cnt'].sum() * 100)) + '%')


### 평균치(6.1)보다 잘하면 승리, 못하면 패배하는 경우가 많고 게임중 승리했을 때의 평균치보다 높을 때 비율이 패배했을 경우보다 높은걸로 보아 내가 잘해야 게임을 이길 수 있는 경우가 많다는 것을 알 수 있음
```
